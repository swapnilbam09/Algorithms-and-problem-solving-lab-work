# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
def dijkstra_enhanced(graph,start):
  distances = {node : float('inf') for node in graph}
  predecessors = {node : None for node in graph}
  visited = []
  distances[start] = 0
  while len(visited) < len(graph) :
    min_node = None
    min_distance = float('inf')
    for node in graph:
      if node not in visited and distances[node] < min_distance:
        min_distance = distances[node]
        min_node = node

    if min_node is None : break
    visited.append(min_node)

    for neighbor, weight in graph[node].items():
      new_distance = distances[min_node] + weight
      if new_distance < distances[neighbor]:
        distances[neighbor] = new_distance
        predecessors[neighbor] = min_node
  return distances,predecessors

def get_path_string(predecessors,start,end):
  path = []
  current = end
  while current is not None:
    path.insert(0,current)
    current = predecessors[current]
    return "->".join(path) if (path and path[0] == start) else "No Path"

def get_user_graph():
  print("\n--- Custom Graph Input ---")
  user_graph = {}
  try:
        num_nodes = int(input("Enter number of nodes: "))
        for _ in range(num_nodes):
            node = input("Node Name: ").strip().upper()
            user_graph[node] = {}
            neighbors = input(f"Enter neighbors and weights for {node} (e.g., B:4,C:2) or 'none': ")
            if neighbors.lower() != 'none':
                for pair in neighbors.split(','):
                    n, w = pair.split(':')
                    user_graph[node][n.strip().upper()] = int(w)
        return user_graph
  except ValueError:
        print("Invalid input format. Using default expanded graph.")
        return None


expanded_graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10, 'E': 3},
    'C': {'A': 2, 'B': 5, 'D': 3, 'F': 8},
    'D': {'B': 10, 'C': 3, 'E': 7, 'F': 1},
    'E': {'B': 3, 'D': 7, 'F': 2},
    'F': {'C': 8, 'D': 1, 'E': 2}
}

choice = input("Do you want to (1) Use the expanded example graph or (2) Enter your own? ")

if choice == '2':
    current_graph = get_user_graph() or expanded_graph
else:
    current_graph = expanded_graph

source = input(f"Enter source node (Options: {', '.join(current_graph.keys())}): ").strip().upper()

if source in current_graph:
    dists, parents = dijkstra_enhanced(current_graph, source)
    print(f"\n--- Shortest Paths from {source} ---")
    for target in current_graph:
        path_str = get_path_string(parents, source, target)
        print(f"To {target}: Distance = {dists[target]} | Path: {path_str}")
else:
    print("Invalid source node.")
```
