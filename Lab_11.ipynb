# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

```python
import sys

def solve_tsp(graph):

    n = len(graph)

    dp = [[-1] * (1 << n) for _ in range(n)]
    parent = [[-1] * (1 << n) for _ in range(n)]

    def visit(city, mask):

        if mask == (1 << n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        best_next_city = -1

        for next_city in range(n):

            if not (mask & (1 << next_city)):
                new_cost = graph[city][next_city] + visit(next_city, mask | (1 << next_city))

                if new_cost < ans:
                    ans = new_cost
                    best_next_city = next_city

        dp[city][mask] = ans
        parent[city][mask] = best_next_city
        return ans


    min_cost = visit(0, 1)

    path = [0]
    curr_city = 0
    curr_mask = 1

    while True:
        next_city = parent[curr_city][curr_mask]
        if next_city == -1: break
        path.append(next_city)
        curr_mask |= (1 << next_city)
        curr_city = next_city

    path.append(0)
    return min_cost, path

def get_user_graph():

    try:
        n = int(input("Enter number of cities: "))
        print(f"Enter the {n}x{n} adjacency matrix (one row per line, space-separated):")
        graph = []
        for i in range(n):
            row = list(map(int, input(f"Row {i+1}: ").split()))
            if len(row) != n:
                print("Error: Row length does not match number of cities.")
                return None
            graph.append(row)
        return graph
    except ValueError:
        print("Error: Please enter valid integers.")
        return None

if __name__ == "__main__":
    print("--- TSP Solver ---")
    print("1. Enter graph manually")
    print("2. Run predefined test cases")
    choice = input("Select an option (1/2): ")

    if choice == '1':
        graph = get_user_graph()
        if graph:
            cost, path = solve_tsp(graph)
            print(f"\nMinimum travelling cost: {cost}")
            print(f"Path taken: {' -> '.join(map(str, path))}")

    elif choice == '2':

        test_graphs = [
            [[0, 10, 15], [10, 0, 20], [15, 20, 0]],
            [[0, 10, 15, 20], [10, 0, 35, 25], [15, 35, 0, 30], [20, 25, 30, 0]]
        ]
        for i, graph in enumerate(test_graphs):
            cost, path = solve_tsp(graph)
            print(f"\nTest Case {i+1} ({len(graph)} cities):")
            print(f"Cost: {cost} | Path: {' -> '.join(map(str, path))}")
```
