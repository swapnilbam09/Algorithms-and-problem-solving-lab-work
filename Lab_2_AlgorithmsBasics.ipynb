```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
t7 = time.time()
# this fucntion calculate the mid of value and check the intersection
print(func4([1,2,3,4],[2,4,6,8]))
t8 = time.time()
print((t8 - t7)*1000)
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
t11 = time.time()
print(func8([3,1,2,7,8,6]))
t12 = time.time()
print((t12 - t11)*1000)
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  s = 0
  for i in arr:
    s += i
  return s
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr2):
  count_even = 0
  count_odd = 0
  for i in arr2:
    if i%2==0:
      count_even += 1
    else:
      count_odd += 1
  return count_even,count_odd
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr3,target):
  for i in arr3:
    if i==target:
      return True
  return False
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def binary_search(arr,target):
  left = 0
  right = len(arr) - 1
  while(left <= right):
    mid = left + (right - left)//2
    if target==arr[mid]:
      return mid
    elif target > arr[mid]:
      left = mid + 1
    else:
      right = mid - 1

  return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
target = int(input("Enter element to search : "))
print("searched element using binary search : ",binary_search([1,3,5,7,9,15,16,14,45],target))
```
